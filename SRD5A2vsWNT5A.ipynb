```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_scatter_from_csv(csv_file, x_col, y_col):
    df = pd.read_csv(csv_file)

    x = df[x_col]
    y = df[y_col]

    plt.scatter(x, y)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f"Correlation between {x_col} and {y_col}")
    plt.show()

plot_scatter_from_csv('/SRD5A2-WNT5A-colab.csv', 'SRD5A2 expression', 'WNT5A expression')
```
